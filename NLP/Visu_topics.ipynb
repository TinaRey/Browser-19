```python
import pandas as pd
import pyLDAvis.gensim as gensimvis
import pyLDAvis
pyLDAvis.enable_notebook()
from os.path import join as pj
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize
import gensim
import nltk
nltk.download('wordnet')
from sklearn.model_selection import train_test_split
np.random.seed(400)
```

```python
def lemmatize_stemming(text):
    return PorterStemmer().stem(WordNetLemmatizer().lemmatize(text, pos='v'))
# Tokenize and lemmatize
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
            
    return result
```

```python
path = 'C:/Users/tr/CodeVsCovid19/data'
file_name = 'metadata_en_topics.csv'
df = pd.read_csv(pj(path, file_name))
```

```python
lda_model = gensim.models.LdaMulticore.load(pj('C:/Users/tr/CodeVsCovid19/models','lda_3.model'))
```

```python
dict_lda = lda_model.id2word
```

```python
processed_docs=[]
for t in df.abstract:
    processed_docs.append(preprocess(t))
    
dictionary = gensim.corpora.Dictionary(processed_docs)
dictionary.filter_extremes(no_below=15, no_above=0.1, keep_n= 100000)
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
vis_data = gensimvis.prepare(lda_model, bow_corpus, dict_lda)
pyLDAvis.display(vis_data)
```
