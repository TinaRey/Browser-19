Goal is to reformat dates that do not have the same format for different sources

if day or month are missing, there will be -1

the dataset must be copied in the data folder

### types of sources

biorxiv (631): 
- year-month-day (*i.e. 2014-01-10*)

Elsevier (18'695): 
- year-month-day(*i.e.1972-12-31*)

WHO (1'102):   
- year (*i.e.2020* - **only for 2020**)
- Nan (9 articles with NaN dates)
- year-month-date (*i.e.2020-02-25* - for some 2020 articles and all other dates - totla 613 articles)

PMC (24'595):   
- year (*i.e. 1989* - **not only for 2020**)
- year-month-date (*i.e. 2008-09-27*)

medrxiv (587): 
- year-month-day (*i.e. 2020-03-24*)

CZI (164):  
- year-month-day (*i.e. 2020-02-28*) (**only 2/164**)
- year (*i.e. 2019*) (**only 2019 2020**)

####  reformat
If missing the day and date I put 0
If NaN then the date will be 0000-00-00 (NaNs can still be found as the only values with the year being 0)

```python
df = metadata_pd
df['publish_time'].fillna('0000-00-00',  inplace=True) #get rid of the nans

df['publish_time'] = df['publish_time'].astype(str)

mask = (~df["publish_time"].astype(str).str.contains('-'))
df.loc[df[mask].index, 'publish_time'] = df['publish_time'].astype(str) + '-00-00'

display(df[df['source_x'] == 'WHO']['publish_time'])
```
