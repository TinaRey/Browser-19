```python
import pandas as pd
import numpy as np
import fasttext.util
```

```python
meta = pd.read_csv("CORD-19-research-challenge/metadata.csv")
```

You can download the model here: https://fasttext.cc/docs/en/language-identification.html#content

```python
ft = fasttext.load_model('lid.176.bin')
```

```python
ft.predict('Der Kuchen ist eine Lüge.')
```

```python
ft.predict(meta.title[4603])[0][0][9:]
```

```python
meta_langue = meta.copy()
```

```python
# New language column
meta_langue['language'] = None
i=0
idx = meta_langue.loc[(meta_langue.abstract.notna())&(meta_langue.abstract!='Unknown')].index
# Fill language column except when there is no abstract or if abstract=='Unknown'
for a in meta_langue.loc[meta_langue.abstract.notna()&(meta_langue.abstract!='Unknown')].abstract:
    meta_langue['language'].loc[idx[i]]=ft.predict(a)[0][0][9:]
    i+=1
```

Look into our new language values:

```python
meta_langue.language.value_counts()
```

```python
# Check for errors
for i in meta_langue.loc[meta_langue.language=='es'].abstract:
    print(i,'\n')
```
