```python
import json
```

```python
filename = '../Covid19_Papers/docs/data.json'
```

```python
file = json.load(open(filename, 'rb'))
```

```python
file
```

```python
nodes, edges = file['nodes'], file['edges']
```

```python
nodes[0]
```

```python
max_x = 0
min_x = 0
max_y = 0
min_y = 0
```

```python
for node in nodes:
    if node['x'] > max_x:
        max_x = node['x']
    if node['x'] < min_x:
        min_x = node['x']
    if node['y'] > max_y:
        max_y = node['y']
    if node['y'] < min_y:
        min_y = node['y']
```
